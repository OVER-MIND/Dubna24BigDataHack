```python
import os
```

```python
# Путь к основной директории, содержащей папки с годами
base_dir = r'all china'
```

```python
# Создаем пустой список для хранения DataFrames
dataframes = []
```

```python
authors = ['Автор (ы)', 'Authors']
article = ['Название документа', 'Title']
source = ['Название источника', 'Source title']
year= ['Год', 'Year']
orgs = ['Организации','Affiliations' ]
auth_orgs = [ 'Авторы организаций','Authors with affiliations']
keys = [ 'Ключевые слова автора','Author Keywords']

# columns = {
#     "authors" : authors,
#     'article' : article,
#     'source' : source,
#     'year' : year,
#     'orgs' : orgs,
#     'keys' : keys
# }
columns = authors + article + source +year + orgs + keys + auth_orgs
rename_dict = {
    'Автор (ы)': 'Authors',
    'Название документа': 'Title',
    'Название источника': 'Source title',
    'Год': 'Year',
    'Организации':'Affiliations',
    'Авторы организаций':'Authors with affiliations',
    'Ключевые слова автора':'Author Keywords'
}
```

```python
from pandas import DataFrame

def filter_data_by_keys(df: DataFrame) -> DataFrame:
    
    new_columns = [ 'Authors','Title','Source title','Year','Affiliations' ,'Authors with affiliations','Author Keywords']
    new_key = 'Author Keywords'
    for column_name in columns:
        # Проверяем, существует ли столбец с таким названием в DataFrame
        if column_name in df.columns:
            # Если 
            if column_name not in new_columns:
                df.rename(columns={ column_name: rename_dict[column_name]}, inplace=True)


    new_df = df[new_columns]
    return new_df[new_df[new_key].str.contains(pattern, case=False, na=False)]
```

```python

# Проходим по всем папкам в указанной директории
for year_folder in os.listdir(base_dir):
    # Полный путь к папке года
    year_path = os.path.join(base_dir, year_folder)
    flag = 0
    # Проверяем, является ли текущий путь директорией
    if os.path.isdir(year_path):
        
        # Проходим по всем файлам в папке года
        for file in os.listdir(year_path):
            # Проверяем, является ли файл CSV
            if file.endswith('.csv'):
                # Полный путь к файлу
                file_path = os.path.join(year_path, file)
                
                # Чтение CSV файла в DataFrame
                df = pd.read_csv(file_path, low_memory=False)
                df = filter_data_by_keys(df)
                # print(df)
                # Добавляем DataFrame в список
                dataframes.append(df)
    

# Объединяем все DataFrames в один
# combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df = pd.concat(dataframes, ignore_index=True).drop_duplicates()
combined_df
```

```python
def SocialRaiting(row):
    authors = row['Authors'].split(";")
    orgs = row['Authors with affiliations'].split(";")
    result_orgs = ""
    result_authors = ""
    for org in orgs:
        if 'China' in org:
            result_orgs+=org +";"
    for name in authors:
        if name in result_orgs:
            result_authors+=name + ";"
    row['Authors'] = result_authors
    row['Authors with affiliations'] = result_orgs
    return row
```

убираем всех не китайцев!)

```python
result_df = combined_df.apply(SocialRaiting, axis=1)
result_df
```
